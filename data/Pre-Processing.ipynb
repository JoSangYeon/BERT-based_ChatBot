```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_excel('Source_data/감성대화말뭉치(Training).xlsx')
valid = pd.read_excel('Source_data/감성대화말뭉치(Validation).xlsx')

train.head()
```

```python
train = train.iloc[:, 7:]
valid = valid.iloc[:, 7:]

valid.head()
```

```python
train_1 = train.iloc[:, :2]
train_2 = train.iloc[:, 2:4]
train_3 = train.iloc[:, 4:6]
train_4 = train.iloc[:, 6:8]

valid_1 = valid.iloc[:, :2]
valid_2 = valid.iloc[:, 2:4]
valid_3 = valid.iloc[:, 4:6]
valid_4 = valid.iloc[:, 6:8]

train_1.columns = ["Q", "A"]
train_2.columns = ["Q", "A"]
train_3.columns = ["Q", "A"]
train_4.columns = ["Q", "A"]

valid_1.columns = ["Q", "A"]
valid_2.columns = ["Q", "A"]
valid_3.columns = ["Q", "A"]
valid_4.columns = ["Q", "A"]

valid_1.head()
```

```python
data = pd.concat([train_1, train_2, train_3, train_4, 
                  valid_1, valid_2, valid_3, valid_4], axis=0)

print("Data's Length :", len(data))
data.head()
```

```python
data = data.dropna()

len(data)
```

```python
data.notnull().sum()
```

```python
data.to_csv('QA_data.csv', index=False)
```
